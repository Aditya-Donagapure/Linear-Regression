## Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
customers.info()
```

```python
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

## Training the mdoel on ulti linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python

lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.score(X, y)
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coef'])
print(cdf)
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X_train)
model = sm.OLS(y_train, X)
model_fit = model.fit()
print(model_fit.summary())
```

## Predicting Test Data

```python
predictions = lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test, y=predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predictions')
plt.title('Yearly Amount Spent vs. Model Predictions')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test, predictions))
print('Mean Squared Error:',mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, predictions)))
```

## Residuals

```python
residuals = y_test-predictions
sns.distplot(residuals, bins=30)
```

```python
import pylab 
import scipy.stats as stats

stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```

### Understanding the Impact:

If the residuals corresponding to -1 and -2 quantiles fall close to the straight line on the plot, it suggests that the residuals in those tails are consistent with what's expected in a normal distribution.
However, if the residuals deviate significantly from the line in the area of negative quantiles, it might indicate that there are more (or less) extreme values in the lower tail of the residuals than expected in a normal distribution. This could be a sign of deviations from normality.
By analyzing how the residuals align with the theoretical quantiles across the entire plot (including negative values), you can gain insights into how well the residuals follow the assumed distribution (typically normal in linear regression).
